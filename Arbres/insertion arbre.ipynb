```python
class Arbre:
    def __init__(self):
        # Le premier élément est la taille de l'arbre
        self._liste = [0]
    
    def _rechercher_indice(self, element):
        indice = 1
        while indice <  len(self._liste) and self._liste[indice] is not None:
            if element < self._liste[indice]:
                indice = indice * 2
            elif element > self._liste[indice]:
                indice = indice * 2 + 1
            else:
                break
                
        return indice
    
    
    def inserer(self, element):
        # précondition : l'élément n'est pas dans l'arbre
        
        indice = self._rechercher_indice(element)
        
        # on créer une longueur de liste suffisante pour acceuillir l'élément        
        if indice >= len(self._liste):
            self._liste += [None] * (indice - len(self._liste) + 1)
            
        self._liste[indice] = element
        # on incrémente la taille
        self._liste[0] += 1
```

```python
OBJETS = {
    0: (100, 5),
    1: (55, 3),
    2: (18, 1),
    3: (70, 4)
}

VALEUR_ACTUELLE = 0
POIDS_LIMITE = 7

def a_rejeter(chemin):
    return sum(OBJETS[objet][1]for objet in chemin) > POIDS_LIMITE

def solution_valide(chemin):
    
    global VALEUR_ACTUELLE
    
    valeur_chemin = sum(OBJETS[objet][0]for objet in chemin)
    if valeur_chemin > VALEUR_ACTUELLE:
        VALEUR_ACTUELLE = valeur_chemin
        return True
    else:
        return False
    
def recherche_solution(chemin=None):
    if chemin is None:
        chemin = []
        
    if a_rejeter(chemin):
        return
    
    if solution_valide(chemin):
        print(chemin)
        
    if len(chemin) == 0:
        noeud_actuel = -1
    else:  
        noeud_actuel = chemin[-1]
    
    for i in range(noeud_actuel + 1, len(OBJETS)):
        recherche_solution(chemin + [i])
```

```python
recherche_solution()
```
